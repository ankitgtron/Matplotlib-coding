```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ['Home Rent', 'Food' ,'Phone/Internet bills', 'Car', 'Other utilities']
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels = exp_labels, radius = 1.5, autopct = '%0.2f%%', explode = [0,0.1,0.1,0.1,0])
plt.savefig("piechart.png", bbox_inches = "tight", pad_inches = 2, transparent = True )
```
