```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113,85,90,150,149,83,98,115,135,80,77,82,129]

plt.hist(blood_sugar, bins = [80,100,125,150], rwidth = 0.95)
```

```python
blood_sugar_men = [113,85,90,150,149,83,98,115,135,80,77,82,129]
blood_sugar_women = [85,90,125,100,142,70,130,80,95, 10, 132, 79, 99]

plt.xlabel("Sugar range")
plt.ylabel("Total no. of patients")
plt.title("Blood sugar analysis ")

plt.hist([blood_sugar_men, blood_sugar_women], bins = [80,100,125,150], rwidth = 0.95,orientation = 'horizontal', color = ['green', 'yellow'],label = ['men', 'women'])
plt.legend()
```
